```python
import sys,json
from typing import Callable, Dict, List, Set, Tuple
from elasticsearch import Elasticsearch
import datetime
sys.path.insert(1, 'util')
from helper_function import preprocess
```

```python
def get_baseline_result(filepath:str,es:Elasticsearch)->Dict[str,List]:
    '''get top30 DBOtype by searching query  in "dbpdiea_type_centric" elasticsearch
    
        returns:dictionary. testing question id ~ top30 DBOtype list
    '''
      
    with open(filepath, 'r') as f:
        smarttask_dbpedia_test = json.load(f)
    print(len(smarttask_dbpedia_test))
    
    map_testID_ranktype={}
    count=0
    print("-----------------",datetime.datetime.now())

    for entry in smarttask_dbpedia_test:

        if entry['question']==None:
            continue

        if entry['category']=='resource':
            
            if count%50==0:
                print("-----------------",datetime.datetime.now())
                print(f'{count} questions has been processed')
            count+=1
            question_processed=preprocess(entry['question'])                         
            hits = es.search(
                index="dbpdiea_type_centric", q=question_processed, _source=True, size=30
            )["hits"]["hits"]
            rank_list= [hit['_source']["type"] for hit in hits]
            #print(rank_list)
            map_testID_ranktype[entry['id']]=rank_list
            #print(map_testID_ranktype)

    return map_testID_ranktype
```

```python
es= Elasticsearch(timeout=600)
es.info()
filepath="../smart-dataset/datasets/DBpedia/smarttask_dbpedia_test.json"
map_testID_ranktype=get_baseline_result(filepath,es)

with open("data/baseline_result.json", 'w') as f:
    json.dump(map_testID_ranktype, f)
print("data/baseline_result.json")
```
